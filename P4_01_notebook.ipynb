# Introduction : Construire un modèle de scoring

"Prêt à dépenser" est une société financière  qui propose des crédits à la consommation pour des personnes ayant peu ou pas d'historique de prêt.

<img src="data/images/pad.png" />

Pour accorder un crédit à la consommation, l’entreprise calcule la probabilité qu’un client le rembourse, ou non. Elle souhaite donc développer un algorithme de scoring pour aider à décider si un prêt peut être accordé à un client.

Les chargés de relation client seront les utilisateurs du modèle de scoring. Puisqu’ils s’adressent aux clients, ils ont besoin que votre modèle soit facilement interprétable. Les chargés de relation souhaitent, en plus, disposer d’une mesure de l’importance des variables qui ont poussé le modèle à donner cette probabilité à un client.

## Descriptions

Huit fichiers de données sont répertoriés et décrits ci-dessous. Notez que l'ensemble de données de l'application a été précédemment divisé en données de formation et de test. 

**application_{train|test}.csv**

- Il s'agit de la table principale, divisée en deux fichiers pour Train (avec TARGET) et Test (sans TARGET). Données statiques pour toutes les applications. Une ligne représente un prêt dans notre échantillon de données. 

**bureau.csv**

- Tous les crédits précédents du client fournis par d'autres institutions financières qui ont été rapportés au Credit Bureau (pour les clients qui ont un prêt dans notre échantillon). Pour chaque prêt de notre échantillon, il y a autant de lignes que le nombre de crédits que le client avait dans le Credit Bureau avant la date de la demande. 

**bureau_balance.csv**

- Soldes mensuels des crédits précédents dans Credit Bureau. Cette table a une rangée pour chaque mois d'historique de chaque crédit précédent rapporté au Credit Bureau - c'est-à-dire que la table a ( nombre de prêts dans l'échantillon x nombre de crédits précédents relatifs x nombre de mois où nous avons un historique observable pour les crédits précédents) rangées. 

**POS_CASH_balance.csv**

- Instantanés des soldes mensuels des prêts POS (point of sale) et cash précédents que le demandeur a eu avec Home Credit. Ce tableau comporte une ligne pour chaque mois de l'historique de chaque crédit précédent au Home Credit (crédit à la consommation et prêts au comptant) lié aux prêts de notre échantillon - c'est-à-dire que le tableau comporte ( nombre de prêts dans l'échantillon x nombre de crédits précédents relatifs x nombre de mois dans lesquels nous avons un historique observable pour les crédits précédents). 

**credit_card_balance.csv**

- Instantanés des soldes mensuels des cartes de crédit précédentes que le demandeur possède chez Home Credit. Cette table a une ligne pour chaque mois de l'historique de chaque crédit précédent dans Home Credit (crédit à la consommation et prêts de trésorerie) lié aux prêts dans notre échantillon - c'est-à-dire que la table a ( nombre de prêts dans l'échantillon x nombre de cartes de crédit précédentes relatives x nombre de mois où nous avons un historique observable pour la carte de crédit précédente) des lignes. 

**previous_application.csv**

- Toutes les demandes précédentes de crédit immobilier des clients qui ont des prêts dans notre échantillon. Il y a une ligne pour chaque demande antérieure liée à des prêts dans notre échantillon de données. 

**installments_payments.csv**

- Historique des remboursements pour les crédits précédemment déboursés dans Home Credit liés aux prêts de notre échantillon. Il y a a) une ligne pour chaque paiement qui a été effectué plus b) une ligne pour chaque paiement manqué. ## Valeurs manquantes

Il est important de traiter les valeurs manquantes dans l'ensemble de données car lorsque les variables sont sélectionnées par IV et VIF, il ne peut y avoir de valeurs manquantes. Pour cette raison, chaque type de valeur sera traité d'une manière différente. 

Lors d'une première tentative, toutes les lignes contenant une valeur NaN ont été supprimées. Si le nombre de lignes supprimées était faible par rapport au nombre total de lignes, cela pourrait être une bonne façon de traiter les valeurs manquantes. Mais dans ce cas, beaucoup de lignes ont des valeurs manquantes, donc cette méthode supprime ~90% du total des lignes. Dans ce projet, chaque colonne sera traitée indépendamment. 

- Catégorielle - NaN remplacé par la valeur catégorielle "unknown" 

- Numérique :

    - AVG - NaN remplacé par la valeur moyenne de la colonne 

    - MEDI - NaN remplacé par la valeur médiane de la colonne. 

    - MODE - NaN remplacé par la valeur du mode de la colonne.

## Traitement des valeurs aberrantes

Les valeurs aberrantes sont définies comme des valeurs supérieures à 3 écarts types par rapport à la moyenne, que ce soit dans le sens positif ou négatif. Ce traitement est appliqué à chaque colonne du DataFrame et uniquement aux types de colonnes numériques. Le traitement des valeurs aberrantes ne doit pas être appliqué aux colonnes de type flag/booléen, c'est-à-dire aux colonnes dont les valeurs sont 0 et 1, car cela fausserait considérablement les données de ces colonnes. 

### Application_test|train

Si une valeur se situe en dehors de 3 écarts-types de la moyenne (6 au total), il faut remplacer la valeur par le 3e écart-type dans l'une ou l'autre direction (ceci peut être fait en deux étapes).

```python
a.describe()
```

```python
# exemple de distribution de données numériques

ax1 = plt.subplot(1, 3, 1)
ax1 = sns.violinplot(y=a.AMT_ANNUITY)

ax2 = plt.subplot(1, 3, 2)
ax2 = sns.violinplot(y=a.AMT_CREDIT)

ax3 = plt.subplot(1, 3, 3)
ax3 = sns.violinplot(y=a.AMT_GOODS_PRICE)

_ = plt.tight_layout()
```

```python
# liste des colonnes numériques
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
num_cols = a.select_dtypes(include=numerics).columns.tolist()
num_cols.remove('TARGET')
num_cols
```

```python
# ne pas appliquer le traitement des valeurs aberrantes
# aux colonnes de type FLAG

flag_cols = []

for col in num_cols:
    vals = a[col].unique().tolist()
    if len(vals) < 3:
        flag_cols.append(col)

num_cols = [x for x in num_cols if x not in flag_cols]
num_cols
```

```python
# traiter les valeurs aberrantes

for col in num_cols:

    # calculer l'écart-type par colonne
    std_dev = a[col].std()

    # calculer la moyenne par colonne
    mean = a[col].mean()

    # calculer le max et le min
    mp = 3
    max = mean + mp * std_dev
    min = mean - mp * std_dev

    # si la colonne est une colonne 'count',
    # alors elle doit être un entier
    # et elle ne peut pas être < 0
    if 'CNT' in col:
        max = round(max, 0)
        if min < 0:
            min = 0.0
        else:
            min = round(min, 0)

    # si la valeur est supérieure à 3 écarts types de la moyenne
    # dans le sens positif, la remplacer par cette valeur
    a.loc[a[col] > max, col] = max

    # si la valeur est supérieure à 3 écarts-types de la moyenne
    # dans le sens négatif, la remplacer par cette valeur
    a.loc[a[col] < min, col] = min
```

```python
# exemple de distribution de données numériques
# après traitement des valeurs aberrantes

ax1 = plt.subplot(1, 3, 1)
ax1 = sns.violinplot(y=a.AMT_ANNUITY)

ax2 = plt.subplot(1, 3, 2)
ax2 = sns.violinplot(y=a.AMT_CREDIT)

ax3 = plt.subplot(1, 3, 3)
ax3 = sns.violinplot(y=a.AMT_GOODS_PRICE)

_ = plt.tight_layout()
```

```python
# traiter les valeurs aberrantes

for col in num_cols:

    # calculer l'écart-type par colonne
    std_dev = b[col].std()

    # calculer la moyenne par colonne
    mean = b[col].mean()

    # calculer le max et le min
    mp = 3
    max = mean + mp * std_dev
    min = mean - mp * std_dev

    # si la colonne est une colonne 'count',
    # alors elle doit être un entier
    # et elle ne peut pas être < 0
    if 'CNT' in col:
        max = round(max, 0)
        if min < 0:
            min = 0.0
        else:
            min = round(min, 0)

    # si la valeur est supérieure à 3 écarts types de la moyenne
    # dans le sens positif, la remplacer par cette valeur
    b.loc[b[col] > max, col] = max

    # si la valeur est supérieure à 3 écarts-types de la moyenne
    # dans le sens négatif, la remplacer par cette valeur
    b.loc[b[col] < min, col] = min
```

```python
# convertir les noms de colonnes en valeurs de lignes
corr_filter = pd.DataFrame(corr.stack())
# convertir l'index multiple en nouvelles colonnes
corr_filter.reset_index(inplace=True)
# définir les noms des colonnes
corr_filter.columns = ['var_1', 'var_2', 'corr']

corr_filter['var_min'] = np.minimum(corr_filter['var_1'], corr_filter['var_2'])
corr_filter['var_max'] = np.maximum(corr_filter['var_1'], corr_filter['var_2'])
corr_filter['var_comb'] = corr_filter['var_min'] + \
    '_+_' + corr_filter['var_max']
corr_filter = corr_filter.drop_duplicates('var_comb')
corr_filter = corr_filter[['var_comb', 'var_1', 'var_2', 'corr']]

corr_filter = corr_filter[corr_filter['var_1'] !=
                          corr_filter['var_2']]  # supprimer la corrélation a:a
filt = 0.995  # définir le filtre min pour la valeur de corrélation
corr_filter = corr_filter[(corr_filter['corr'] > filt)
                          | (corr_filter['corr'] <
                             -1 * filt)]  # appliquer le filtre

corr_filter['corr_abs'] = corr_filter['corr'].abs()
corr_filter = corr_filter.sort_values(by=['corr_abs'])
corr_filter = corr_filter.drop('corr_abs', axis=1)

corr_filter
```

## Encodage

L'encodage est un processus dans lequel une colonne unique est développée en de nouvelles colonnes afin de simplifier le dataframe. Nous verrons deux types d'encodage : le Label Encoding et le One-Hot-Encoding. 

- **Label Encoding** : Attribue à chaque catégorie unique d'une variable catégorielle un nombre entier. Aucune nouvelle colonne n'est créée. A utiliser pour les colonnes avec seulement 2 catégories ou une où les catégories sont ordonnées : (ex. : tailles = petites, moyennes, grandes).

<img src="data/images/le.png"/>

- **One-Hot-Encoding** : Crée une nouvelle colonne pour chaque catégorie unique dans une variable catégorielle. Chaque observation reçoit un 1 dans la colonne pour sa catégorie correspondante et un 0 dans toutes les autres nouvelles colonnes. A utiliser pour les colonnes avec 3+ catégories et les données catégorielles où l'ordre n'a pas d'importance : (ex. : couleurs = rouge, bleu, vert). 

<img src="data/images/ohe.png"/>

### Analyse de la valeur de l'information (Information Value ou IV)

L'analyse de la valeur de l'information (Information Value ou IV) est une technique d'exploration de données qui permet de déterminer quelles colonnes de données d'un ensemble de données ont un pouvoir prédictif ou une influence sur la valeur d'une variable dépendante spécifiée.

```python
# Importation de bibliothèques pour l'IV

from scipy import stats
import pandas.core.algorithms as algos
import re
import traceback
```

```python
max_bin = 20
force_bin = 3
```

```python
# définir une fonction de binning
def mono_bin(Y, X, n=max_bin):

    df1 = pd.DataFrame({"X": X, "Y": Y})
    justmiss = df1[['X', 'Y']][df1.X.isnull()]
    notmiss = df1[['X', 'Y']][df1.X.notnull()]
    r = 0
    while np.abs(r) < 1:
        try:
            d1 = pd.DataFrame({
                "X": notmiss.X,
                "Y": notmiss.Y,
                "Bucket": pd.qcut(notmiss.X, n)
            })
            d2 = d1.groupby('Bucket', as_index=True)
            r, p = stats.spearmanr(d2.mean().X, d2.mean().Y)
            n = n - 1
        except Exception as e:
            n = n - 1

    if len(d2) == 1:
        n = force_bin
        bins = algos.quantile(notmiss.X, np.linspace(0, 1, n))
        if len(np.unique(bins)) == 2:
            bins = np.insert(bins, 0, 1)
            bins[1] = bins[1] - (bins[1] / 2)
        d1 = pd.DataFrame({
            "X":
            notmiss.X,
            "Y":
            notmiss.Y,
            "Bucket":
            pd.cut(notmiss.X, np.unique(bins), include_lowest=True)
        })
        d2 = d1.groupby('Bucket', as_index=True)

    d3 = pd.DataFrame({}, index=[])
    d3["MIN_VALUE"] = d2.min().X
    d3["MAX_VALUE"] = d2.max().X
    d3["COUNT"] = d2.count().Y
    d3["EVENT"] = d2.sum().Y
    d3["NONEVENT"] = d2.count().Y - d2.sum().Y
    d3 = d3.reset_index(drop=True)

    if len(justmiss.index) > 0:
        d4 = pd.DataFrame({'MIN_VALUE': np.nan}, index=[0])
        d4["MAX_VALUE"] = np.nan
        d4["COUNT"] = justmiss.count().Y
        d4["EVENT"] = justmiss.sum().Y
        d4["NONEVENT"] = justmiss.count().Y - justmiss.sum().Y
        d3 = d3.append(d4, ignore_index=True)

    d3["EVENT_RATE"] = d3.EVENT / d3.COUNT
    d3["NON_EVENT_RATE"] = d3.NONEVENT / d3.COUNT
    d3["DIST_EVENT"] = d3.EVENT / d3.sum().EVENT
    d3["DIST_NON_EVENT"] = d3.NONEVENT / d3.sum().NONEVENT
    d3["WOE"] = np.log(d3.DIST_EVENT / d3.DIST_NON_EVENT)
    d3["IV"] = (d3.DIST_EVENT-d3.DIST_NON_EVENT) * \
        np.log(d3.DIST_EVENT/d3.DIST_NON_EVENT)
    d3["VAR_NAME"] = "VAR"
    d3 = d3[[
        'VAR_NAME', 'MIN_VALUE', 'MAX_VALUE', 'COUNT', 'EVENT', 'EVENT_RATE',
        'NONEVENT', 'NON_EVENT_RATE', 'DIST_EVENT', 'DIST_NON_EVENT', 'WOE',
        'IV'
    ]]
    d3 = d3.replace([np.inf, -np.inf], 0)
    d3.IV = d3.IV.sum()

    return (d3)
```

```python
def char_bin(Y, X):

    df1 = pd.DataFrame({"X": X, "Y": Y})
    justmiss = df1[['X', 'Y']][df1.X.isnull()]
    notmiss = df1[['X', 'Y']][df1.X.notnull()]
    df2 = notmiss.groupby('X', as_index=True)

    d3 = pd.DataFrame({}, index=[])
    d3["COUNT"] = df2.count().Y
    d3["MIN_VALUE"] = df2.sum().Y.index
    d3["MAX_VALUE"] = d3["MIN_VALUE"]
    d3["EVENT"] = df2.sum().Y
    d3["NONEVENT"] = df2.count().Y - df2.sum().Y

    if len(justmiss.index) > 0:
        d4 = pd.DataFrame({'MIN_VALUE': np.nan}, index=[0])
        d4["MAX_VALUE"] = np.nan
        d4["COUNT"] = justmiss.count().Y
        d4["EVENT"] = justmiss.sum().Y
        d4["NONEVENT"] = justmiss.count().Y - justmiss.sum().Y
        d3 = d3.append(d4, ignore_index=True)

    d3["EVENT_RATE"] = d3.EVENT / d3.COUNT
    d3["NON_EVENT_RATE"] = d3.NONEVENT / d3.COUNT
    d3["DIST_EVENT"] = d3.EVENT / d3.sum().EVENT
    d3["DIST_NON_EVENT"] = d3.NONEVENT / d3.sum().NONEVENT
    d3["WOE"] = np.log(d3.DIST_EVENT / d3.DIST_NON_EVENT)
    d3["IV"] = (d3.DIST_EVENT-d3.DIST_NON_EVENT) * \
        np.log(d3.DIST_EVENT/d3.DIST_NON_EVENT)
    d3["VAR_NAME"] = "VAR"
    d3 = d3[[
        'VAR_NAME', 'MIN_VALUE', 'MAX_VALUE', 'COUNT', 'EVENT', 'EVENT_RATE',
        'NONEVENT', 'NON_EVENT_RATE', 'DIST_EVENT', 'DIST_NON_EVENT', 'WOE',
        'IV'
    ]]
    d3 = d3.replace([np.inf, -np.inf], 0)
    d3.IV = d3.IV.sum()
    d3 = d3.reset_index(drop=True)

    return (d3)
```

```python
def data_vars(df1, target):

    stack = traceback.extract_stack()
    filename, lineno, function_name, code = stack[-2]
    vars_name = re.compile(r'\((.*?)\).*$').search(code).groups()[0]
    final = (re.findall(r"[\w']+", vars_name))[-1]

    x = df1.dtypes.index
    count = -1

    for i in x:
        if i.upper() not in (final.upper()):
            if np.issubdtype(df1[i],
                             np.number) and len(pd.Series.unique(df1[i])) > 2:
                conv = mono_bin(target, df1[i])
                conv["VAR_NAME"] = i
                count = count + 1
            else:
                conv = char_bin(target, df1[i])
                conv["VAR_NAME"] = i
                count = count + 1

            if count == 0:
                iv_df = conv
            else:
                iv_df = iv_df.append(conv, ignore_index=True)

    iv = pd.DataFrame({'IV': iv_df.groupby('VAR_NAME').IV.max()})
    iv = iv.reset_index()
    return (iv_df, iv)
```

### Facteur d'inflation de la variance (Variance Inflation Factor ou VIF)

Le facteur d'inflation de la variance (Variance Inflation Factor ou VIF) est le rapport de la variance d'un modèle comportant plusieurs termes, divisé par la variance d'un modèle comportant un seul terme. Il quantifie la gravité de la multicollinéarité dans une analyse de régression par moindres carrés ordinaires.

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def iterate_vif(df, vif_threshold=5, max_vif=6):
    count = 0
    while max_vif > vif_threshold:
        count += 1
        print("Itération # " + str(count))
        vif = pd.DataFrame()
        vif["VIFactor"] = [
            variance_inflation_factor(df.values, i) for i in range(df.shape[1])
        ]
        vif["features"] = df.columns

        if vif['VIFactor'].max().round(1) > vif_threshold:
            print('Suppression de %s avec VIF de %f' %
                  (vif[vif['VIFactor'] == vif['VIFactor'].max()]
                   ['features'].values[0], vif['VIFactor'].max().round(1)))
            df = df.drop(vif[vif['VIFactor'] == vif['VIFactor'].max()]
                         ['features'].values[0],
                         axis=1)
            max_vif = vif['VIFactor'].max().round(1)
        else:
            print('Complété')
            return df, vif.sort_values('VIFactor').round(1)
```

###  Fonction d'entraînement, de prédiction et de score

```python
def train_and_score_model(model_name,
                          estimator,
                          param_grid=None,
                          cv=5,
                          iid=False):

    # Étape 5 : Instancier l'objet CV
    if model_name in ['Modèle factice', 'Classification naïve bayésienne']:
        cv = estimator
    elif model_name == 'Régression logistique':
        cv = GridSearchCV(estimator, param_grid, cv=cv, iid=False)
    else:
        cv = RandomizedSearchCV(estimator,
                                param_grid,
                                cv=cv,
                                iid=False,
                                random_state=None)

    # Étape 6 : Ajustement de l'entraînement
    t0 = time.time()
    cv.fit(X_train, y_train)
    ttt = time.time(
    ) - t0  # ttt pour temps d\'apprentissage, "time to train" en anglais
    print("Il faut %.3f secondes pour que %s s'ajuste" % (ttt, model_name))
    metrics.loc['temps d\'apprentissage', model_name] = ttt

    # Étape 7 : Prédiction sur le test.
    y_pred = cv.predict(X_test)

    # Étape 8 : Scoring
    # Accuracy

    print("L'accuracy avec \"" + model_name +
          "\" sur l'ensemble du test est de %.3f" % cv.score(X_test, y_test))

    metrics.loc['accuracy', model_name] = accuracy_score(y_pred=y_pred,
                                                         y_true=y_test)
    metrics.loc['precision', model_name] = precision_score(y_pred=y_pred,
                                                           y_true=y_test)
    metrics.loc['recall', model_name] = recall_score(y_pred=y_pred,
                                                     y_true=y_test)

    if model_name != 'Machine à vecteurs de support':
        y_pred_prob = cv.predict_proba(X_test)[:, 1]
        metrics.loc['auc', model_name] = roc_auc_score(y_test, y_pred_prob)

    print(Conf_Matrix(confusion_matrix(y_test, y_pred)))
    print(classification_report(y_test, y_pred))

    if model_name != 'Machine à vecteurs de support':
        # ROC Curve
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
        plt.plot([0, 1], [0, 1], 'k--')
        plt.plot(fpr, tpr, label=model_name)
        plt.xlabel('Taux de faux positifs')
        plt.ylabel('Taux de vrais positifs')
        plt.title('%s Courbe ROC' % model_name)
        plt.show()

    print('_____________________________________________')

    return cv
```

```python
def Conf_Matrix(CM, labels=[0, 1]):
    df = pd.DataFrame(data=CM, index=labels, columns=labels)
    df.index.name = 'VRAI'
    df.columns.name = 'PRÉDICTION'
    df.loc['Total'] = df.sum()
    df['Total'] = df.sum(axis=1)
    return df
```
